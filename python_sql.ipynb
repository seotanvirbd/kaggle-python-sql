```python
#Not avilable and "unknown" datas are turned into NaN that means na_values
df1 = pd.read_csv('orders.csv',na_values =['Not Available', 'unknown'])
df1['Ship Mode'].unique()
```

```python
df1[df1['Ship Mode'].isna()==1] 
```

```python
#rename columns names ..make them lower case and replace space with underscore
df1.rename(columns={'Order Id':'order_id', 'City':'city'})
```

```python
# another good way to rename the columns
df1.columns=df1.columns.str.lower()
df1.columns=df1.columns.str.replace(' ','_')
```

```python
df1
```

```python
#derive new columns discount from the multiplication of column  sale price and profit
df1['discount']= df1['list_price']*df1['discount_percent']*.01
```

```python
# create sale_price column
df1['sale_price']=df1['list_price']- df1['discount']
```

```python
#create profit column
df1['profit'] = df1['sale_price']- df1['cost_price']
df1
```

```python
#check thr data type of the columns
df1.dtypes
```

```python
#convert order date from object data type to datetime
df1['order_date']=pd.to_datetime(df1['order_date'],format="%Y-%m-%d")
df1.head()
```

```python
df1.dtypes
```

```python
#drop cost price list price and discount percent columns
df1.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
df1.head()
```
